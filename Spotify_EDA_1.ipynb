```python
import os

# Path to the folder containing your CSV files
folder_path = 'archive (27)'

# Get a list of all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Create an empty list to store individual DataFrames
dfs = []

# Loop through the CSV files and read them
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    dfs.append(df)

# Concatenate all DataFrames
combined_df = pd.concat(dfs, ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('indian_songs_combined.csv', index=False)

print("All files have been successfully combined into 'indian_songs_combined.csv'")
```

```python
df = pd.read_csv('indian_songs_combined.csv')
df
```

```python
from sklearn.utils import shuffle
df = shuffle(df)
df
```

```python
print(df.isnull().sum())
```

```python
df_filled = df.fillna(df.mean(numeric_only=True))
```

```python
df_filled
```

```python
print(df_filled.isnull().sum())
```

```python
# Separate numerical and categorical columns
numerical_features = df_filled.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = df_filled.select_dtypes(include=['object', 'category']).columns.tolist()

# Print the column names
print("Numerical Features:")
print(numerical_features)
print("\nCategorical Features:")
print(categorical_features)
```

```python
# Descriptive statistics
desc_stats = df_filled[numerical_features].describe()
print("\nDescriptive Statistics:\n", desc_stats)
```

```python
# Top 10 singers
top_singers = df_filled['singer'].value_counts().head(10)
print("\nTop 10 Singers:\n", top_singers)
```

```python
# Top 5 languages
top_languages = df_filled['language'].value_counts().head(5)
print("\nTop 5 Languages:\n", top_languages)
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_filled[numerical_features].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Heatmap of Numerical Features")
plt.tight_layout()
plt.show()
```

```python
# Distribution of numerical features
for col in numerical_features:
    plt.figure(figsize=(6, 4))
    sns.histplot(df_filled[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```

```python
# Count plot for top 5 languages
plt.figure(figsize=(8, 4))
sns.countplot(data=df_filled[df_filled['language'].isin(top_languages.index)], x='language', order=top_languages.index)
plt.title("Top 5 Languages by Song Count")
plt.xlabel("Language")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Top streamed songs
top_streamed = df_filled[['song_name', 'Stream', 'language']].sort_values(by='Stream', ascending=False).head(10)
print("\nTop 10 Streamed Songs:\n", top_streamed)
```

```python
# Popularity vs Stream
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_filled, x='popularity', y='Stream', hue='language', alpha=0.6)
plt.title("Popularity vs Streams by Language")
plt.tight_layout()
plt.show()
```

```python
# Convert released_date to datetime and extract year
df_filled['release_year'] = pd.to_datetime(df_filled['released_date'], errors='coerce').dt.year
songs_by_year = df_filled['release_year'].value_counts().sort_index()

plt.figure(figsize=(12, 5))
songs_by_year.plot(kind='bar')
plt.title("Number of Songs Released per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Average danceability and energy over years
trend = df_filled.groupby('release_year')[['danceability', 'energy']].mean().dropna()
trend.plot(figsize=(10, 5))
plt.title("Trends of Danceability and Energy Over the Years")
plt.ylabel("Average Value")
plt.tight_layout()
plt.show()
```

```python
# Outlier Detection using IQR for selected features
outlier_features = ['Stream', 'popularity', 'tempo', 'loudness']
for col in outlier_features:
    Q1 = df_filled[col].quantile(0.25)
    Q3 = df_filled[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_filled[(df_filled[col] < lower_bound) | (df_filled[col] > upper_bound)]
    print(f"\nNumber of outliers in {col}: {len(outliers)}")
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df_filled[col])
    plt.title(f"Boxplot for {col} (Outliers included)")
    plt.tight_layout()
    plt.show()
```

```python
# Popular Songs Trend over Time
popular_songs = df_filled[df_filled['popularity'] > 80]
popular_by_year = popular_songs['release_year'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
popular_by_year.plot(kind='line', marker='o')
plt.title("Count of Popular Songs (popularity > 80) Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Popular Songs")
plt.tight_layout()
plt.show()
```

```python
# Stream Trend by Year
avg_streams_by_year = df_filled.groupby('release_year')['Stream'].mean().dropna()

plt.figure(figsize=(10, 5))
avg_streams_by_year.plot(marker='o')
plt.title("Average Streams per Song Over the Years")
plt.xlabel("Year")
plt.ylabel("Average Streams")
plt.tight_layout()
plt.show()
```

```python
# Duet vs Solo vs Group Analysis
df_filled['num_singers'] = df_filled['singer'].apply(lambda x: len(str(x).split('|')))
df_filled['singer_type'] = df_filled['num_singers'].apply(lambda x: 'Solo' if x == 1 else ('Duet' if x == 2 else 'Group'))

plt.figure(figsize=(6, 4))
sns.countplot(data=df_filled, x='singer_type')
plt.title("Distribution of Solo, Duet, and Group Songs")
plt.xlabel("Singer Type")
plt.ylabel("Number of Songs")
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Clustering Songs by Numerical Features
features_for_clustering = ['danceability', 'acousticness', 'energy', 'liveness', 'loudness', 'speechiness', 'tempo', 'Valence']
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_filled[features_for_clustering])

kmeans = KMeans(n_clusters=4, random_state=42)
df_filled['cluster'] = kmeans.fit_predict(scaled_features)

# PCA for 2D visualization
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)
df_filled['pca1'] = pca_result[:, 0]
df_filled['pca2'] = pca_result[:, 1]

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_filled, x='pca1', y='pca2', hue='cluster', palette='Set2')
plt.title("KMeans Clustering of Songs (PCA Projection)")
plt.tight_layout()
plt.show()
```

```python
df_filled = df_filled.drop(columns=['singer_id', 'duration', 'released_date' ])
df_filled
```
